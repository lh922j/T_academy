```python
import requests
from bs4 import BeautifulSoup
import time
import re

result = []
base_url = 'http://ufcstats.com/statistics/events/completed?page=all'
res = requests.get(base_url)
soup = BeautifulSoup(res.text, "html.parser")
events = soup.find("tbody").find_all("tr")

for i, event in enumerate(events[2:], start=1):  # 2번 인덱스부터 시작
    try:
        print(f"[{i}] 이벤트 페이지 접근 중...")
        event_url = event.find("a").get("href")
        res_event = requests.get(event_url)
        soup_event = BeautifulSoup(res_event.text, "html.parser")
        time.sleep(1)

        matches = soup_event.find("tbody").find_all("tr")
        for j, match in enumerate(matches):
            try:
                fight_url = match.find_all("a")[0].get("href")
                res_fight = requests.get(fight_url)
                soup_fight = BeautifulSoup(res_fight.text, "html.parser")
                print(f"  - [{i}번째 이벤트] {j+1}번째 경기 접근 중")
                time.sleep(1)

                rounds = soup_fight.find_all("tbody")[1].find_all("th")
                for k, round_header in enumerate(rounds):
                    try:
                        print(f"    > {k+1}라운드 수집 중...")

                        # 경기 정보
                        f1 = soup_fight.find_all("tbody")[0].find_all("a")[0].text
                        f2 = soup_fight.find_all("tbody")[0].find_all("a")[1].text
                        referee = soup_fight.find_all("div")[10].find_all("span")[0].text.strip()
                        re_1 = re.sub(r'\s+', ' ', soup_fight.find_all("div")[10].find_all('i')[14].text.strip()).replace(' - ', '-').replace(' .', '.')
                        re_2 = re.sub(r'\s+', ' ', soup_fight.find_all("div")[10].find_all('i')[15].text.strip()).replace(' - ', '-').replace(' .', '.')
                        re_3 = re.sub(r'\s+', ' ', soup_fight.find_all("div")[10].find_all('i')[16].text.strip()).replace(' - ', '-').replace(' .', '.')
                        method = soup_fight.find_all("div")[10].find_all('i')[3].text.strip()
                        r1 = round_header.text.strip()

                        # f1 데이터
                        f1_kd1 = soup_fight.find_all("tbody")[1].find_all("tr")[k].find_all("p")[2].text.strip()
                        f1_total1 = soup_fight.find_all("tbody")[1].find_all("td")[4 + 10*k].find_all("p")[0].text.strip()
                        f1_td1 = soup_fight.find_all("tbody")[1].find_all("td")[5 + 10*k].find_all("p")[0].text.strip()
                        f1_tdp1 = soup_fight.find_all("tbody")[1].find_all("td")[6 + 10*k].find_all("p")[0].text.strip()
                        f1_ctrl1 = soup_fight.find_all("tbody")[1].find_all("td")[9 + 10*k].find_all("p")[0].text.strip()
                        f1_sig1 = soup_fight.find_all("tbody")[3].find_all("td")[1 + 9*k].find_all("p")[0].text.strip()
                        f1_head1 = soup_fight.find_all("tbody")[3].find_all("td")[3 + 9*k].find_all("p")[0].text.strip()
                        f1_body1 = soup_fight.find_all("tbody")[3].find_all("td")[4 + 9*k].find_all("p")[0].text.strip()
                        f1_leg1 = soup_fight.find_all("tbody")[3].find_all("td")[5 + 9*k].find_all("p")[0].text.strip()
                        f1_dis1 = soup_fight.find_all("tbody")[3].find_all("td")[6 + 9*k].find_all("p")[0].text.strip()
                        f1_cli1 = soup_fight.find_all("tbody")[3].find_all("td")[7 + 9*k].find_all("p")[0].text.strip()
                        f1_ground1 = soup_fight.find_all("tbody")[3].find_all("td")[8 + 9*k].find_all("p")[0].text.strip()

                        # f2 데이터
                        f2_kd1 = soup_fight.find_all("tbody")[1].find_all("tr")[k].find_all("p")[3].text.strip()
                        f2_total1 = soup_fight.find_all("tbody")[1].find_all("td")[4 + 10*k].find_all("p")[1].text.strip()
                        f2_td1 = soup_fight.find_all("tbody")[1].find_all("td")[5 + 10*k].find_all("p")[1].text.strip()
                        f2_tdp1 = soup_fight.find_all("tbody")[1].find_all("td")[6 + 10*k].find_all("p")[1].text.strip()
                        f2_ctrl1 = soup_fight.find_all("tbody")[1].find_all("td")[9 + 10*k].find_all("p")[1].text.strip()
                        f2_sig1 = soup_fight.find_all("tbody")[3].find_all("td")[1 + 9*k].find_all("p")[1].text.strip()
                        f2_head1 = soup_fight.find_all("tbody")[3].find_all("td")[3 + 9*k].find_all("p")[1].text.strip()
                        f2_body1 = soup_fight.find_all("tbody")[3].find_all("td")[4 + 9*k].find_all("p")[1].text.strip()
                        f2_leg1 = soup_fight.find_all("tbody")[3].find_all("td")[5 + 9*k].find_all("p")[1].text.strip()
                        f2_dis1 = soup_fight.find_all("tbody")[3].find_all("td")[6 + 9*k].find_all("p")[1].text.strip()
                        f2_cli1 = soup_fight.find_all("tbody")[3].find_all("td")[7 + 9*k].find_all("p")[1].text.strip()
                        f2_ground1 = soup_fight.find_all("tbody")[3].find_all("td")[8 + 9*k].find_all("p")[1].text.strip()
                        
                        result.append([
                            r1, referee, method, re_1, re_2, re_3,
                            f1, f1_kd1, f1_total1, f1_kd1, f1_td1, f1_tdp1, f1_ctrl1,
                            f1_sig1, f1_head1, f1_body1, f1_leg1, f1_dis1, f1_cli1,
                            f2, f2_kd1, f2_total1, f2_kd1, f2_td1, f2_tdp1, f2_ctrl1,
                            f2_sig1, f2_head1, f2_body1, f2_leg1, f2_dis1, f2_cli1
                        ])
                    except Exception as e:
                        print(f"라운드 수집 실패 [round {k+1}]: {e}")
            except Exception as e:
                print(f" 경기 접근 실패 [match {j+1}]: {e}")
    except Exception as e:
        print(f"이벤트 접근 실패 [event {i}]: {e}")
        continue

print(f"총 {len(result)}개의 라운드 데이터를 수집했습니다.")
```
